```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

### Q1
Read in the files gold.txt and labels.txt.

The gold DataFrame should have columns url and category, while the labels DataFrame should have columns turk, url and category.

```python
#

labels = pd.read_csv('labels.txt', header= None, delimiter="\t")
labels.columns = ['turks', 'url', 'category']

labels
```

```python
gold = pd.read_csv('gold.txt', header= None, delimiter="\t")
gold.columns = ['url', 'category']

gold
```

### Q2

Split the labels DataFrame into two: labels on gold and labels unknown


```python
labels_on_gold = labels[labels['url'].isin(gold['url'])]
labels_unknown = labels[~(labels['url'].isin(gold['url']))]

labels_on_gold
labels_unknown
```

### Q3
Create a rater goodness DataFrame that is indexed by turk, and has two
columns: the number of ratings, and the average correctness of ratings for
each turk (both on gold set urls)

```python
#creating separate columns where url and category is combined to compare on both
labels_on_gold['c'] = labels_on_gold['url'] + ' ' + labels_on_gold['category']
gold['c'] = gold['url'] + ' ' + gold['category']

#sets the labels on gold column correct to 1 if it predicts the right category
labels_on_gold['correct'] = 0
(labels_on_gold.loc[labels_on_gold['c'].isin(gold['c']), ['correct']]) = 1

rater_goodness = labels_on_gold.groupby('turks')['category'].count()
rater_goodness = pd.DataFrame(rater_goodness)
rater_goodness.columns=['num_of_ratings']
#calculates the number correct
rater_goodness['num_correct'] = labels_on_gold.groupby('turks')['correct'].sum()
#calculates the num correct over the total number for accuracy
rater_goodness['accuracy'] = (rater_goodness['num_correct'])/rater_goodness['num_of_ratings']
rater_goodness = rater_goodness.drop(columns='num_correct')

(rater_goodness)
```

### Q4
Attach a column called odds to the rater goodness DataFrame, using the
average correctness of the turk as his or her p.

```python
rater_goodness['odds'] = rater_goodness['accuracy']/(1.001 - rater_goodness['accuracy'])
rater_goodness
```

### Q5
List the top 10 most accurate turks who have rated at least 20 gold set URLs.


```python
mask = rater_goodness['num_of_ratings'] > 19
rater_goodness[mask].sort_values('accuracy', ascending = False)[0:10]
```

### Q6
Is number of ratings by a turker on gold set URLs related to his or her
accuracy?

```python
import statsmodels.api as sm
from patsy import dmatrices

y, X = dmatrices('accuracy ~ num_of_ratings', rater_goodness, return_type='dataframe')
result = sm.OLS(y, X).fit()      # Fit model (find the intercept and slopes)
print result.summary()

```

From this regression summary we see that the p-value for number of ratings in regards to accuracy is well above 0.05, thus we can say that number of ratings is not significant in prediciting accuracy. We can also look at the R^2 value which is very close to 0 suggesting that there is little to no correlation.

```python
slope = result.params['num_of_ratings']
intercept = result.params['Intercept']
print 'accuracy = {:.4f} + {:.4f} * num_of_ratings'.format(intercept, slope)

predicted = rater_goodness['num_of_ratings'] * slope + intercept
regression_predictions = Series(predicted.values,
                                index=rater_goodness['num_of_ratings'])
regression_predictions[:5]
```

```python
plot(rater_goodness['num_of_ratings'], rater_goodness['accuracy'], marker='o', linestyle='None')

# New plot command for the regression predictions
regression_predictions.plot(label='Regression', linewidth=2)

xlabel('number of ratings')
ylabel('accuracy')
legend(numpoints=1, loc='best')
show()
```

From this graph we see the plot of the ratings against the accuracy as well as the regression plot. Just looking at the scatterplot there doesn't seem to be any correlation and the line as well misses most of the points.

### Q7
For the pair (u, c), calculate the product of odds of all turks who (a) rated url u as
category c, and (b) have rated more gold set urls than 75% of all turks.

```python
#set top_25 to be the 75% quantile of the rater goodness dataframe
rater_goodness =rater_goodness.sort_values('num_of_ratings', ascending = False)
top_25 = rater_goodness[rater_goodness['num_of_ratings'] > rater_goodness['num_of_ratings'].quantile(0.75)]
```

```python
labels_unknown3 = labels_unknown[['url','turks', 'category']]
labels_unknown3['top25'] = 0
(labels_unknown3.loc[labels_unknown3['turks'].isin(top_25.index.tolist()), ['top25']]) = 1
labels_unknown3 = labels_unknown3[labels_unknown3['top25'] ==1]
#labels_unknown3 #determines if the turk that rated url u with category c is in the top 25% of turks who rated gold urls

rater_split = rater_goodness
rater_split = rater_split.reset_index()[['turks','odds']]
lu_df = labels_unknown3.merge(rater_split,
                             left_on='turks',
                             right_on='turks')
odds_df = lu_df.groupby(['url','category'])['odds'].prod() #multiplies the odds of the turks together based on (u,c)
odds_df = odds_df.reset_index()
odds_df.rename(index=str,columns={'odds': 'overall odds'}, inplace = True)
odds_df
```

### Q8

Create a DataFrame (called result 75, whose index is URLs not in the gold
set, and with two columns called top category and top odds. The top
category should be the category with the highest overall odds for that url,
and the top odds should be the overall odds for that top category

```python
result_75 = odds_df
result_75 = result_75.set_index('url')
#groupby url then calculate max overall odds. for each url, keep the row in which overall odds == max
result_75 = (result_75[result_75.groupby('url')['overall odds'].transform(max) == result_75['overall odds']]) #groups by url and takes the max of the options
result_75.columns=['top category','top odds'] #change column names
result_75 = result_75[~result_75.index.duplicated(keep='first')] #some urls may have a tie for max so drop one of them
result_75
```

### Q9

Repeat the code of Q7 and Q8, but replacing 75% by 25% in the description
of Q7 (i.e., we also consider turks who have far fewer gold set ratings).
Call this result 25.

Create a DataFrame where
both the index and the columns are the various categories, and the cells
contain the number of urls with these as the top categories according to
result 75 and result 25

Where are the most errors?


```python
#top_75 is the 25% quantile of rater goodness
top_75 = rater_goodness[rater_goodness['num_of_ratings'] > rater_goodness['num_of_ratings'].quantile(0.25)]
labels_unknown4 = labels_unknown[['url','turks', 'category']]
labels_unknown4['top75'] = 0
(labels_unknown4.loc[labels_unknown4['turks'].isin(top_75.index.tolist()), ['top75']]) = 1
#labels_unknown4 #determines if the turk that rated url u with category c is in the top 25% of turks who rated gold urls
labels_unknown4 = labels_unknown4[labels_unknown4['top75'] ==1]
rater_split2 = rater_goodness
rater_split2 = rater_split2.reset_index()[['turks','odds']]
lu_df2 = labels_unknown4.merge(rater_split2,
                             left_on='turks',
                             right_on='turks')
odds_df2 = lu_df2.groupby(['url','category'])['odds'].prod()
odds_df2 = odds_df2.reset_index()
odds_df2.rename(index=str,columns={'odds': 'overall odds'}, inplace = True)
result_25 = odds_df2
result_25 = result_25.set_index('url')
#groupby url then calculate max overall odds. for each url, keep the row in which overall odds == max
result_25 = (result_25[result_25.groupby('url')['overall odds'].transform(max) == result_25['overall odds']]) #groups by url and takes the max of the options
result_25.columns=['top category_25','top odds_25'] #change column names
result_25 = result_25[~result_25.index.duplicated(keep='first')] #some urls may have a tie for max so drop one of them
result_25 #calculates like in Q8 to the top category and odds for 25% quantile
```

```python
#creates crosstab of result_75 and result_25
#top category is for the 75 percentile and above
#top category_25 is the 25 percentile and above
cross = pd.crosstab(result_75['top category'], result_25['top category_25'])
cross
```

The most errors occurred when the url was rated P by the top raters and not P by the others. About 43% of the ratings that were supposed to be P were incorrect. Counts wise, the most errors occured in G as there were 976 errors in ratings. 
But we can see that the majority per each category were rated correctly.
